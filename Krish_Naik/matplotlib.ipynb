## Pipe Chart

```python
labels = ['A', 'B', 'C', 'D'] 
sizes = [30, 20, 40, 10]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.2, 0, 0, 0)

## create apie chart
plt.pie(sizes, labels = labels, colors = colors, autopct = '%1.1f%%', shadow = True)
```

```python
labels = ['A', 'B', 'C', 'D'] 
sizes = [30, 20, 40, 10]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.2, 0, 0, 0) ## move out the first slice

## create apie chart
plt.pie(sizes, explode = explode, labels = labels, colors = colors, autopct = '%1.1f%%', shadow = True)
```

```python
## Sales Data Visulisation
import pandas as pd
sales_data_df = pd.read_csv('sales_data.csv')
sales_data_df.head(5)
```

```python
# Plot Total Sale by Product
total_sales_by_product = sales_data_df.groupby('Product')['Total_Sale'].sum()
print(total_sales_by_product)
```

```python
total_sales_by_product.plot(kind = 'bar', color = 'teal')
```

```python
## plot sales trend over time
sales_data_df.groupby('Date')['Total_Sale'].sum()
```

```python
sales_data_df.groupby('Date')['Total_Sale'].sum().reset_index()
```

```python
sales_trend = sales_data_df.groupby('Date', as_index=False)['Total_Sale'].sum()

plt.plot(sales_trend['Date'], sales_trend['Total_Sale'])
```
