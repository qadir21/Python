```python
ob1 = pd.DataFrame({'A' : [1, 2, 3, 4, 5], 'B' : [6, 7, 8, 9, 10]})
ob2 = pd.DataFrame({'A' : [1, 2, 3, 4, 5], 'C' : [16, 17, 18, 19, 20]})
pd.merge(ob1, ob2, on = 'A')
```

```python
ob1 = pd.DataFrame({'A' : [1, 2, 3, 4, 5], 'B' : [6, 7, 8, 9, 10]})
ob2 = pd.DataFrame({'A' : [1, 2, 3, 4, 6], 'C' : [16, 17, 18, 19, 20]})
pd.merge(ob1, ob2, how = 'left')
```

```python
ob1 = pd.DataFrame({'A' : [1, 2, 3, 4, 5], 'B' : [6, 7, 8, 9, 10]})
ob2 = pd.DataFrame({'A' : [1, 2, 3, 4, 6], 'C' : [16, 17, 18, 19, 20]})
pd.merge(ob1, ob2, how = 'right')
```

```python
ob1 = pd.DataFrame({'A' : [1, 2, 3, 4, 5], 'B' : [6, 7, 8, 9, 10]})
ob2 = pd.DataFrame({'A' : [1, 2, 3, 4, 6], 'C' : [16, 17, 18, 19, 20]})
pd.merge(ob1, ob2, how = 'outer')
```
